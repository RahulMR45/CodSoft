```python
class contact_book:
    contact = []

    def add(self, l):
        self.contact.append(l)

    def disp(self):
        for i in self.contact:
            print("name", i[0])
            print("phone", i[1])
            print("email", i[2])
            print("address", i[3])
            print()

    def search(self, m):
        if self.contact == []:
            print("contact list is empty")
        for i in self.contact:
            if m == i[0] or m == i[1]:
                print("name", i[0])
                print("phone", i[1])
                print("email", i[2])
                print("address", i[3])
                return
        print("contact does not exists")

    def update(self, m, l):

        for i in self.contact:
            if m == i[0] or m == i[1]:
                if l[0].strip():
                    i[0] = l[0]
                if l[1].strip():
                    i[1] = l[1]
                if l[2].strip():
                    i[2] = l[2]
                if l[3].strip():
                    i[3] = l[3]
            print("Contact updated successfully:")
            print("name", i[0])
            print("phone", i[1])
            print("email", i[2])
            print("address", i[3])
            return

        print("Contact does not exist")

    def delete(self, m):
        for i in self.contact:
            if m == i[0] or m == i[1]:
                self.contact.remove(i)
                print("Contact deleted successfully")
                return
        print("contact not exists")
        return


```

```python
x = contact_book()
while True:
    print("\nContact Book Menu")
    print("1. Add Contact")
    print("2. View Contact List")
    print("3. Search Contact")
    print("4. Update Contact")
    print("5. Delete Contact")
    print("6. Exit")
    c = int(input("Enter your choice: "))

    if c == 1:
        l = []
        l.append(input("Enter name: "))
        l.append(input("Enter phone: "))
        l.append(input("Enter email: "))
        l.append(input("Enter address: "))
        x.add(l)
        print("Contact added successfully")
    elif c == 2:
        x.disp()
        print("Contact displayed successfully")
    elif c == 3:
        m = input("enter the name or phone number:")
        x.search(m)
        print("Contact searched successfully")
    elif c == 4:
        l = []
        m = input("Enter name or phone for updating:")
        l.append(input("Enter new name(blank to keep current):"))
        l.append(input("Enter new phone(blank to keep current):"))
        l.append(input("Enter new email(blank to keep current):"))
        l.append(input("Enter new address(blank to keep current):"))
        x.update(m, l)
        print("Contact updated successfully")
    elif c == 5:
        m = input("Enter name or phone for updating:")
        x.delete(m)
        print("Contact deleted successfully")
    elif c == 6:
        break
    else:
        print("enter valid choice")
```
